```python
import brian2 as bb
from brian2.units import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as ft
%matplotlib inline
```

```python
def visualize_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    bb.figure(figsize=(10, 4))
    bb.subplot(121)
    bb.plot(bb.zeros(Ns), bb.arange(Ns), 'ok', ms=10)
    bb.plot(bb.ones(Nt), bb.arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        bb.plot([0, 1], [i, j], '-k')
    bb.xticks([0, 1], ['Source', 'Target'])
    bb.ylabel('Neuron index')
    bb.xlim(-0.1, 1.1)
    bb.ylim(-1, max(Ns, Nt))
    bb.subplot(122)
    bb.plot(S.i, S.j, 'ok')
    bb.xlim(-1, Ns)
    bb.ylim(-1, Nt)
    bb.xlabel('Source neuron index')
    bb.ylabel('Target neuron index')
```

```python
plot = True

simulation_duration = 50*ms
num_inputs = 10

input_rates = np.random.randint(50, 150, num_inputs)
input_rates = [item*Hz for item in input_rates]

weight_low_bound = -0.01; weight_high_bound = 0.03;
input_weights = np.random.randint(weight_low_bound*10_000, weight_high_bound*10_000, num_inputs)/10_000

if plot:
    fig, axs = plt.subplots(1,2)
    axs[0].hist([item/Hz for item in input_rates], bins='auto'); axs[0].set_title('Frequency Distribution')
    axs[1].hist(input_weights, bins='auto'); axs[1].set_title('Weight Distribution')

input_weights
```

```python
bb.start_scope()

taum = 10*ms
taupre = 20*ms
taupost = taupre
Ee = 0*mV
vt = -54*mV
vr = -60*mV
El = -74*mV
taue = 5*ms
gmax = weight_high_bound
dApre = 0.01*weight_high_bound
dApost = -dApre * taupre / taupost * 1.05

InputGroup = bb.NeuronGroup(num_inputs, 
                            '''dv/dt=I*Hz : 1
                            I : 1
                            ''', threshold='v>0.1', reset='v=0', method='euler')
InputGroup.I = [item/Hz for item in input_rates]

OutputNeuron = bb.NeuronGroup(1, '''dv/dt = (ge * (Ee-v) + El - v) / taum : volt
                            dge/dt = -ge / taue : 1''',
                            threshold='v>vt', reset='v = vr', method='euler')

Synapses = bb.Synapses(InputGroup, OutputNeuron,  '''w : 1
                dApre/dt = -Apre / taupre : 1 (event-driven)
                dApost/dt = -Apost / taupost : 1 (event-driven)''', 
            on_pre='''ge += w
                    Apre += dApre
                    w = clip(w + Apost, weight_low_bound, gmax)''',
             on_post='''Apost += dApost
                     w = clip(w + Apre, weight_low_bound, gmax)''',
             )
Synapses.connect()
Synapses.w = input_weights
```

```python
StateMonitor = bb.StateMonitor(OutputNeuron, 'v', record=0)
#InputStateMonitor = bb.StateMonitor(InputGroup, 'v', record=list(range(num_inputs)))
InSpikeMonitor = bb.SpikeMonitor(InputGroup)
OutSpikeMonitor = bb.SpikeMonitor(OutputNeuron)
WeightMonitor = bb.StateMonitor(Synapses, 'w', record=Synapses[:,0])
```
