```python
class FullyConnected2DLayer(nn.Module):
    def __init__(self, input_size, output_size, sparsity_index=None, weights=None):
        super(FullyConnected2DLayer, self).__init__()
        self.input_size = input_size
        self.output_size = output_size
        self.input_size_flat = input_size[0] * input_size[1]
        self.output_size_flat = output_size[0] * output_size[1]

        # Initialize weights
        if weights is not None:
            self.weights = nn.Parameter(weights)
        else:
            self.weights = nn.Parameter(torch.randn(self.output_size_flat, self.input_size_flat))
            if sparsity_index is not None:
                self.sparsify_myself(sparsity_index)

    def forward(self, input):
        # Assuming input is of shape (batch_size, input_size[0], input_size[1])
        batch_size = input.shape[0]
        flat_input = input.view(batch_size, self.input_size_flat)
        flat_output = torch.matmul(flat_input, self.weights.t())
        output = flat_output.view(batch_size, *self.output_size)
        return output

    def sparsify_myself(self, sparsity_index):
        with torch.no_grad():
            mask = torch.bernoulli(torch.full(self.weights.size(), sparsity_index))
            self.weights.data.mul_(mask)
            self.weights.data.add_((1 - mask) * 0.001)
            self.weights.data.div_(torch.sqrt(torch.tensor(sparsity_index)))
```

```python
class SimpleSNN(nn.Module):
    def __init__(self, layer_sizes, sparsity_index=None, threshold=1.0):
        self.layer_sizes = layer_sizes
        super(SimpleSNN, self).__init__()
        self.layers = nn.ModuleList()
        self.leakys = nn.ModuleList()

        # Create layers and LIF objects for each layer
        for i in range(len(layer_sizes) - 1):
            input_size, output_size = layer_sizes[i], layer_sizes[i + 1]
            self.layers.append(FullyConnected2DLayer(input_size, output_size, sparsity_index))
            self.leakys.append(snn.Leaky(beta=0.5, threshold=threshold, reset_mechanism='zero'))
        
    def forward(self, input_sequence):
        e = torch.exp(torch.tensor(-1.0))
        spike_records = []
        mem = [lif.init_leaky() for lif in self.leakys]
        voltage_records = [torch.zeros(input_sequence.size(0)+1, *layer_size) for layer_size in self.layer_sizes]

        for t in range(input_sequence.size(0)):
            input_t = input_sequence[t]
            x = input_t
            voltage_records[0][t+1] = e*voltage_records[0][t]+input_t

            for i, (fc, lif) in enumerate(zip(self.layers, self.leakys)):
                x = fc(x)
                voltage_records[i+1][t+1] = e*voltage_records[i+1][t]+x
                spk, mem[i] = lif(x, mem[i])
                x = spk  # Output of one layer is input to the next

            spike_records.append(x)

        output_spikes = torch.stack(spike_records)
        return torch.stack(voltage_records), output_spikes.float()

    def visualize_network(self):
        """
        Visualize the network connections based on the actual size matrix provided during initialization.
        For each layer defined in layer_sizes, it visualizes the neurons and connections between them.
        """
        G = nx.DiGraph()

        # Create a mapping for neurons to their positions
        pos = {}
        layer_positions = []
        current_pos_x = 0  # Initial position for the first layer

        # Add nodes for each layer
        for layer_index, layer_size in enumerate(self.layer_sizes):
            neurons = [f"Layer{layer_index}_Neuron_{i}" for i in range(layer_size[0] * layer_size[1])]
            G.add_nodes_from(neurons)

            # Calculate positions of neurons for this layer
            for i, neuron in enumerate(neurons):
                pos[neuron] = (current_pos_x, -i)  # Negative to have the first layer at the top
                layer_positions.append((current_pos_x, -i))

            current_pos_x += 1  # Increment position for the next layer

        # Add edges with weights between layers
        for layer_index, fc_layer in enumerate(self.layers):
            weight_matrix = fc_layer.weights.data.numpy()
            input_neurons = [f"Layer{layer_index}_Neuron_{i}" for i in range(self.layer_sizes[layer_index][0] * self.layer_sizes[layer_index][1])]
            output_neurons = [f"Layer{layer_index+1}_Neuron_{i}" for i in range(self.layer_sizes[layer_index+1][0] * self.layer_sizes[layer_index+1][1])]

            # Connect each neuron from one layer to each neuron in the next
            for i, input_neuron in enumerate(input_neurons):
                for j, output_neuron in enumerate(output_neurons):
                    weight = weight_matrix[j, i]
                    if weight != 0:
                        G.add_edge(input_neuron, output_neuron, weight=abs(weight))

        # Draw the graph
        edge_widths = [G[u][v]['weight'] * 10 for u, v in G.edges]  # Scale edge thickness by weight for visibility
        nx.draw(G, pos, with_labels=True, node_color="skyblue", width=edge_widths, arrows=True)
        plt.title("Neural Network Connections")
        plt.xlabel("Layers")
        plt.ylabel("Neurons")
        plt.gca().invert_yaxis()  # Invert y-axis to have the first layer at the top in the plot
        plt.show()
```

```python
import torch

def sparse_binary_vector(length, input_dims, p):
    """
    Generate a sparse binary vector of given length and input dimensions,
    where each element has a probability p of being 1.

    :param length: The length of the vector.
    :param input_dims: A tuple representing the dimensions of each element in the vector.
    :param p: The probability of an element being 1.
    :return: A torch tensor of shape [length] + list(input_dims) filled with 0s and 1s.
    """
    # Create a probability tensor of the desired shape filled with p
    prob_tensor = torch.full((length,) + input_dims, p)
    
    # Generate the binary vector with elements drawn from a Bernoulli distribution
    binary_vector = torch.bernoulli(prob_tensor).to(torch.int).unsqueeze(1).float()
    
    return binary_vector.float()
```

```python
import random
from torch.utils.data import DataLoader, Subset
import tonic
import tonic.transforms as transforms
from tonic import datasets
from tonic import DiskCachedDataset
from IPython.display import HTML

# Load the dataset
dataset = datasets.DVSGesture(save_to='./data', train=True)
sensor_size = datasets.DVSGesture.sensor_size

# Define frame transformation
frame_transform = transforms.Compose([
    transforms.CropTime(max=2_000_000),
    transforms.Downsample(spatial_factor=0.25),  # Make sure this downsamples to (32, 32)
    transforms.ToFrame(sensor_size=(32, 32), time_window=10_000)
])

# Load the train and test datasets with the specified transformations
trainset = datasets.DVSGesture(save_to='./data', transform=frame_transform, train=True)
testset = datasets.DVSGesture(save_to='./data', transform=frame_transform, train=False)

# Subsampling parameters
batch_size = 8
subsample_percent = 10

# Subsample datasets
train_subsample_idx = random.sample(range(len(trainset)), int(len(trainset) * subsample_percent // 100))
test_subsample_idx = random.sample(range(len(testset)), int(len(testset) * subsample_percent // 100))
trainset = Subset(trainset, train_subsample_idx)
testset = Subset(testset, test_subsample_idx)

# Cache datasets
cached_trainset = DiskCachedDataset(trainset, cache_path='./cache/DVSGesture/train')
cached_testset = DiskCachedDataset(testset, cache_path='./cache/DVSGesture/test')

# DataLoader
train_loader = DataLoader(cached_trainset, batch_size=batch_size, collate_fn=tonic.collation.PadTensors(batch_first=True), shuffle=True)
test_loader = DataLoader(cached_testset, batch_size=batch_size, collate_fn=tonic.collation.PadTensors(batch_first=True))

# Retrieve a batch of frames and move time to the first dimension
frames, targets = next(iter(train_loader))
frames = frames.movedim(0, 1)

# Check the shape of the frames to ensure they are correct
print(frames.shape)
ani = tonic.utils.plot_animation(frames[:,0])
HTML(ani.to_jshtml())
```
