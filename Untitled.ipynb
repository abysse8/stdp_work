## vvvvv Utilities vvvvv

```python
def add_to_class(Class):
    def wrapper(obj):
        setattr(Class, obj.__name__, obj)
    return wrapper

class SpikeAnimation():
    def __init__(self, num_plots, color_map = 'coolwarm', frames = None):
        self.to_plot = []
        self.sizes = []
        self.ims = []
        self.color_map = color_map
        self.calc_frames = True
        self.fig, self.axs = plt.subplots(1, num_plots)
        if frames is not None:
            self.frames = frames
            self.calc_frames = False
        return
            
                       
    def add_to_plot(self, three_dim_tensor, size, title):
        assert len(three_dim_tensor.shape) == 3, "Not a three dimensional tensor"
        self.to_plot.append(three_dim_tensor.squeeze(1).detach().numpy())
        self.axs[len(self.to_plot)].set_title(title)
        self.sizes.append(size)
        self.ims.append( self.axs[len(self.to_plot)-1].imshow(self.to_plot[-1][0].reshape(size), 
                                      cmap = self.color_map) )
        return
        
    def blit(self, n):
        for iii, image in enumerate(self.ims):
            image.set_array(self.to_plot[iii][n].reshape(self.sizes[iii]))
        return self.ims
    
    def show(self):
        assert len(self.to_plot) != 0, "No spikes loaded"
        ani = matplotlib.animation.FuncAnimation(self.fig, self.blit, frames = self.frames)
        return HTML(ani.to_jshtml())
    
```

## Building the network (no stdp)

```python
class snn_hmax(nn.Module):
    def __init__(self,
                input_shape = (28,28),
                s1_shape = (10,10),
                num_outputs = 10,
                kernel_size = 2,
                stride = None,
                padding = 0,
                num_steps = 25,
                beta = 0.95):
        
        super().__init__()
        
        for passed_parameter in set(locals().keys())-{'self', '__class__'}: #save all arguments as self.xxx
            setattr(self, passed_parameter, locals()[passed_parameter])

        self.num_inputs = input_shape[0]*input_shape[1]
        self.n_neurons1 = s1_shape[0]*s1_shape[1]
        
        self.s1 = snn.Leaky(beta=self.beta)
        self.c1 = snn.Leaky(beta=self.beta)
        self.s2 = snn.Leaky(beta=self.beta)
        self.c2 = snn.Leaky(beta=self.beta)
        
        # "S1 cells detect edges, C1 maps subsample S1 maps (thru square maxpool)
        # S2 cells detect mode complex features, C2 cells select **most salient S2 response**
        # i.e select the **kernels**
        self.input_to_s1 = nn.Conv2d(in_channels=200, out_channels=1, 
                                     kernel_size=3, stride=1, padding=0, groups=1, bias=True, dtype=None)
        self.maxpool1 = nn.MaxPool2d(kernel_size, stride=stride, padding=padding)
        self.s1_to_c1 = nn.Linear(int((self.n_neurons1-self.kernel_size)/
                                  self.kernel_size+1), num_outputs)
        #self.c1_to_s2 = nn.Linear(num_outputs, num_inputs)
        
model = snn_hmax()
```

```python
@add_to_class(snn_hmax)
def getHFIRST(self, spikes_in):
    # Converts three dimensional (2D image+time) array and outputs a 2D map
    # of the first spikes to arrive, HFIRST style. Used in forward pass
    space_dimension = spikes_in.shape[2]
    time_dimension = spikes_in.shape[0]
    h_first_layer = torch.zeros(space_dimension)
    for iii in range(space_dimension):
        spikes_at_node = np.where(spikes_in[:, 0, iii] == 1)[0]
        index_of_first_spike = 0 if len(spikes_at_node)==0 else time_dimension-min(spikes_at_node)
        h_first_layer[iii] = index_of_first_spike
    calculable = h_first_layer.reshape(self.s1_shape[0], self.s1_shape[1]).unsqueeze(0)
    return calculable
```

```python
@add_to_class(snn_hmax)
def forward(self, spikes_in):
    
    s1_init_voltages = self.s1.init_leaky()
    
    # Record the final layer
    s1_spikes_records = []
    s1_voltages_records= []

    for iii in range(self.num_steps):
        spikes_in = spikes_in.reshape(-1, 28, 28)
        print(spikes_in)
        s1_currents = self.input_to_s1(spikes_in)
        x = nn.Linear(28*28, 100)
        spikes_in = spikes_in[iii].reshape(-1, 784)
        s1_currents = x(spikes_in)
        print(s1_currents)
        s1_spikes, s1_voltages = self.s1(s1_currents, s1_init_voltages)
        s1_spikes_records.append(s1_spikes)
        s1_voltages_records.append(s1_voltages)
        
    print(s1_spikes.shape)

    c2_inputs=self.getHFIRST(s1_spikes)
    c2_filtered = self.maxpool1(c2_inputs)

    #plt.imshow(c2_filtered[0])
        
    return torch.stack(s1_spikes_records, dim=0), torch.stack(s1_voltages_records, dim=0), c2_inputs, c2_filtered
    
def __call__(self, spikes_in):
    return self.forward(spikes_in)
    

spikes_in = spikegen.rate_conv(torch.rand((200, 784))).unsqueeze(1)
spikes_out, _ , hfirst, maxpool_output= model(spikes_in)
x = hfirst
hfirst = hfirst.expand(200, 10, 10)
maxpool_output = maxpool_output.expand(200, 5, 5)
```
