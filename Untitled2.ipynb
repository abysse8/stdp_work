```python
import brian2 as bb
from brian2.units import *
import numpy as np
%matplotlib inline
```

```python
def visualize_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    bb.figure(figsize=(10, 4))
    bb.subplot(121)
    bb.plot(bb.zeros(Ns), bb.arange(Ns), 'ok', ms=10)
    bb.plot(bb.ones(Nt), bb.arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        bb.plot([0, 1], [i, j], '-k')
    bb.xticks([0, 1], ['Source', 'Target'])
    bb.ylabel('Neuron index')
    bb.xlim(-0.1, 1.1)
    bb.ylim(-1, max(Ns, Nt))
    bb.subplot(122)
    bb.plot(S.i, S.j, 'ok')
    bb.xlim(-1, Ns)
    bb.ylim(-1, Nt)
    bb.xlabel('Source neuron index')
    bb.ylabel('Target neuron index')
```

```python
bb.start_scope()
bb.defaultclock.dt = 1 * ms

eqs = """dv/dt = (0.04*v**2 + 5*v + 140 - u + I + I_noise )/ms : 1
         du/dt = (a*(b*v - u))/ms  : 1
         I : 1
         I_noise : 1
         a : 1
         b : 1
         c : 1
         d : 1
       """
re = np.random.uniform(size=6)
G = bb.NeuronGroup(6, eqs, threshold="v>=1", reset="v=c; u+=d", method="euler")
G.v = -65

G.I[3:] = [0]*3
G.I_noise = [0]*6

M = bb.StateMonitor(G, 'v', record=[3,4,5])

G.a = 0.02
G.b = 0.2
G.c = -65 + 15 * re**2
G.d = 8 - 6 * re**2
G.u = "b*v"

S = bb.Synapses(G, G, 
                "w : 1",
                on_pre={"up": "I += w", "down": "I -= w"},
                delay={"up": 0 * ms, "down": 1 * ms})
S.connect(i=[0,1,2]*3, j=[3,3,3,4,4,4,5,5,5])
G.I[0:3] = [50, 30, 15]
S.w = [10, 15, 20, 20, 30, 40, 5, 25, 55]
S.w = [iii/2 for iii in S.w]

spike = bb.SpikeMonitor(G)
visualize_connectivity(S)
bb.store()
```
